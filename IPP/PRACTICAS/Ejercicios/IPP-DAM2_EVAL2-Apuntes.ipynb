```python
def procesar_numeros(num1, num2, num3):
    """
    Esta función toma tres numeros com entrada e imprime:
    - Si el primero es positivo:
      - La suma de los tres números
      - O si no, el producto de los tres
    - Si el primero es par:
      - La suma del primero con el producto del segundo y tercero
      - El producto del primero y el segundo sumado al tercero
    """
    if num1 < 0:
        print(f"La multiplicación de los tres números es: {num1 * num2 * num3}")
    else:
        print(f"La suma de los tres números es: {num1 + num2 + num3}")
    if num1 % 2 == 0:
        print(f"La primero sumado al producto de los últimos es: {num1 + num2 * num3}")
    else:
        print(f"El producto de los primeros sumado al último es: {num1 * num2 + num3}")

# Ejemplo de uso de la función
numero1 = float(input("Introduce el primer número: "))
numero2 = float(input("Introduce el segundo número: "))
numero3 = float(input("Introduce el tercer número: "))
procesar_numeros(numero1 ,numero2, numero3)
```

```python
def saludar(nombre, saludo='Buenos dias'):
    """
    Esta función saluda a la persona con un mensaje personalizable
    """
    mensaje = f"{saludo}, {nombre}."
    return mensaje

# Ejemplos de uso de la función
saludo_por_defecto = saludar('Ana')
print(saludo_por_defecto)
saludo_casual = saludar('Mariano', saludo='Hola')
print(saludo_casual)
```

```python
def calcular_factorial(n):
    """
    Esta función calcula el factorial de un número y lo devuelve
    """
    if n < 0:
        return -1
    
    factorial = 1
    if n == 0 || n == 1:
        return factorial
    
    for i in range(2, n + 1):
        factorial = factorial * i
    return factorial

# Ejemplo de uso
numero = int(input("Introduce un número: "))
resultado = calcular_factorial(numero)
if resultado == -1:
    print("No se pudo calcular el factorial del número. El número es negativo.")
else:
    print(resultado)
```
