```python
dia = int(input("Introduce el número del dia de la semana (1-lunes, 7-domingo): "))
match dia:
    case 1:
        print("Lunes")
    case 2:
        print("Martes")
    case 3:
        print("Miércoles")
    case 4:
        print("Jueves")
    case 5:
        print("Viernes")
    case 6:
        print("Sábado")
    case 7:
        print("Domingo")
    case _:
        print("Error. El numero debe de estar entre 1 y 7.")

# Se pueden combinar varios litarales en un solo aptron usando |
error = int(input("Introduce el número de error: "))
match error:
    case 404 | 405 | 406:
        print("Acceso no permititdo.")
```

```python
# Ejemplo de validación
numero = int(input("Introduce un número entre 0 y 10:"))
while numero < 0 or 10 < numero:
    print("Error. El número tiene que estar entre 0 y 10.")
    numero = int(input("Introduce un número entre 0 y 10:"))
print(numero)   
```

```python
# Ejemplo de validación
numero = int(input("Introduce un número que no esté entre 0 y 10:"))
while 0 <= numero and numero <= 10:
    print("Error. El número no puede estar entre 0 y 10.")
    numero = int(input("Introduce un número que no esté entre 0 y 10:"))
print(numero)   
```
