```python
# Algunos ejemplos
var1 = 345
print("var1:", var1, type(var1))

#El tipo de una variable se puede cambiar, pero no se recomienda
var1 = "hola"
print("var1 con diferente tipo:", var1, type(var1))
print("var1 es de tipo str (string): ", isinstance(var1, str))

var2 = None
print("var2:", var2)
```

```python
# En Python no hay constantes
# Si queremos indicar que una variable no debe cambiarse, debemos usar mayúsculas
VENTA_MAX = 4_567
print(VENTA_MAX)

VENTA_MAX = "4"
print(VENTA_MAX)
```

```python
import math

# Algunos ejemplos
var1 = 345_678_935
print("var1:", var1, type(var1))

var2 = "buenos dias"
type(var2)
print("var2:", var2, type(var2))

var3 = 4.5
print("var3:", var3, type(var3))

var4 = 2 + 5j
print("var4:", var4, type(var4))

var5 = math.inf
print("var5:", var5, type(var5))

var6 = math.nan
print("var6:", var6, type(var6))

#Otra anotaciones
print(4_567)
print(1e+6)
print(4_567.4)
print(1e-4)

# Problema de precisión en número flotante
var7 = 9876543210987643.0
print(var7)
```
